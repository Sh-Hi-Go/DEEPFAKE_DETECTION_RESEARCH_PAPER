```python
import cv2
from retinaface import RetinaFace
import os
from torchvision import transforms
import torch
```

```python
# Path to the Celeb-DF dataset (replace with your actual path)
data_dir = "/content/drive/MyDrive/celebDF_dataset/celebDF/Celeb-real"

# Initialize RetinaFace model
#retinaface = RetinaFace()
```

```python
# Loop through subfolders (original and manipulated)
#for subfolder in ["original", "manipulated"]:
subfolder_path = data_dir #os.path.join(data_dir, "")

  # Loop through each video file
for filename in os.listdir(subfolder_path):
  if filename.endswith(".mp4"):
    video_path = os.path.join(subfolder_path, filename)
    cap = cv2.VideoCapture(video_path)

    # Read the first frame
    ret, frame = cap.read()

    if ret:
      # Convert frame to RGB (RetinaFace expects RGB)
      frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

      # Detect faces using RetinaFace
      faces = RetinaFace.extract_faces(frame)
      counter=0
      for face in faces:
        transform = transforms.Compose([transforms.ToPILImage(), transforms.Resize((224, 224)), transforms.ToTensor(), transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
        processed_frame = transform(face)
        # Assuming you have a directory to save processed images
        output_path = os.path.join("/content/drive/MyDrive/CelebDF_processed/celeb_real", f"{filename[:-4]}_{counter}.pt")  # Remove extension and add .pt
        #cv2.imwrite(output_path, processed_frame)
        torch.save(processed_frame, output_path)
        counter = counter+1

    cap.release()
```
