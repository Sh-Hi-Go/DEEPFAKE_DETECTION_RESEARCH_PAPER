```python
from retinaface import RetinaFace
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
from PIL import Image
```

```python
fake_video_path = '/content/drive/MyDrive/celebDF/Celeb-synthesis'
real_video_path = '/content/drive/MyDrive/celebDF/Celeb-real'
fake_img_path = '/content/drive/MyDrive/dataset/fake'   #where fake face image from random frame will be stored
real_img_path = '/content/drive/MyDrive/dataset/real'   #where real face image from random frame will be stored
```

```python
def save_faces_as_images(faces, output_folder,filename):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    # Save each detected face as an image
    for i, face in enumerate(faces):
        # Create a file path for the saved face image
        output_path = os.path.join(output_folder, filename[:-4]+str(i)+".jpg")
        # Save the face as an image
        cv2.imwrite(output_path, cv2.cvtColor(np.array(face), cv2.COLOR_RGB2BGR))
```

```python
import random
def capture_random_frame(path,video, saveAt):
    video_path = os.path.join(path,video)
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open the video file.")
        return
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if(total_frames<130):
      return
    random_frame_number = random.randint(130, total_frames-1)
    cap.set(cv2.CAP_PROP_POS_FRAMES, random_frame_number)
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read the random frame.")
        cap.release()
        return
    faces = RetinaFace.extract_faces(frame)
    save_faces_as_images(faces, saveAt,video)
    cap.release()
```

```python
for video in os.listdir(real_video_path):
  capture_random_frame(real_video_path,video,real_img_path)
```

```python
for video in os.listdir(fake_video_path):
  capture_random_frame(fake_video_path,video,fake_img_path)
```
